# 6. Functional-genomics target validation

Open Targets druggability tractability plus DepMap CRISPR gene essentiality, scored to a verdict. For a loss-of-function driver the harness pivots to the druggable induced dependency.

```python
import os, sys
def _find_repo():
    p = os.getcwd()
    for _ in range(6):
        if os.path.exists(os.path.join(p, "anveshar", "__init__.py")):
            return p
        p = os.path.dirname(p)
    return None
REPO = _find_repo()
if REPO and REPO not in sys.path:
    sys.path.insert(0, REPO)
import anveshar
print("anveshar loaded from", os.path.dirname(anveshar.__file__))
```

## Validate the driver and its induced dependency (live)

```python
from anveshar import pipeline
import pandas as pd
try:
    d = pipeline.run("renal medullary carcinoma", live=True)
    tv = d["target_validation"]
    display(pd.DataFrame([{"gene": g, "score": v.get("score"), "tractability_SM": v.get("sm_tractability"),
                           "common_essential": v.get("common_essential"), "selective_dependency": v.get("selective_dependency"),
                           "verdict": v.get("verdict")} for g, v in tv.items() if v.get("available")]))
    print("Induced dependency:", {g: e["dependency"] for g, e in d.get("induced_dependencies", {}).items()})
except Exception as ex:
    print("live validation skipped:", type(ex).__name__, ex)
```

SMARCB1 is a common-essential tumor suppressor (low score, not druggable directly); the harness hands off to its induced synthetic-lethal dependency EZH2, which validates as a tractable, selective target.
